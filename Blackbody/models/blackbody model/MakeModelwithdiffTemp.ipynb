```python
with open("template.in") as f:
    template = f.read()

for h in range(4,13):
    for t in range(5,8):

        T = 10**t

        text = template.replace("hden 10, power =-2",f"hden {h}, power =-2")

        text = text.replace("blackbody 10000 K",f"blackbody {T} K")

        text = text.replace('file="last_cont.dat"',f'file="last_cont_h{h}_t{T}.dat"')

        with open(f"model_h{h}_t{T}.in", "w") as f:
            f.write(text)

print("All models generated.")
```
