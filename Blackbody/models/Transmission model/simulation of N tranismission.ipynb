```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {}

for i in range(1, 6):
    data[i] = pd.read_csv(f"last_cont_trans{i}.dat", sep="\t")
```

```python
for i in range(1,6):
    cont = data[i]
    energy = cont["#Cont  nu"].to_numpy()
    flux = cont["incident"].to_numpy()

    plt.figure()
    plt.loglog(energy, flux)

    plt.xlabel("Energy in rydbergs")
    plt.ylabel("Flux")

    plt.ylim(1e29,1e40)
    plt.xlim(1e-2,1e4)
    plt.title(f" Incident continuum (transmission number= {i})")
    plt.show()
```

```python

df = pd.read_csv(f"last_cont_h8_t10000.dat", sep="\t")
e_init = df["#Cont  nu"].to_numpy()
init = df["incident"].to_numpy()
for i in range(1,6):
    cont = data[i]
    energy = cont["#Cont  nu"].to_numpy()
    incident  = cont["incident"].to_numpy()
    transmitted = cont["trans"].to_numpy()
    
    plt.figure()
    plt.loglog(e_init, init, label = "initial BB continuum")
    plt.loglog(energy, incident, label ="incident continuum")
    plt.loglog(energy, transmitted, label = "transmitted continuum")

    plt.ylim(1e32,1e40)
    plt.xlim(1e-2,1e3)

    plt.ylabel("flux")
    plt.xlabel("wavelength in microns")
    plt.title(f" Incident and Transmitted continuum (transmission number = {i})")
    plt.legend()
    plt.show()
```

```python
for i in range(1,6):
    cont = data[i]
    energy = cont["#Cont  nu"].to_numpy()
    incident  = cont["incident"].to_numpy()
    net_transmitted = cont["net trans"].to_numpy()

    plt.figure()

    plt.loglog(energy, incident, label ="incident continuum")
    plt.loglog(energy, net_transmitted, label = "net transmitted continuum")

    plt.ylim(1e32,1e40)
    plt.xlim(1e-2,1e3)

    plt.ylabel("flux")
    plt.xlabel("wavelength in microns")
    plt.title(f" Incident and Net Transmitted continuum for transmission = {i}")
    plt.legend()
    plt.show()
```
