```python
import os
```

```python
N = int(input("Write the number of times you want to run the transmit:"))
with open("new_test.in") as f:
    template = f.read()

for n in range(1, N+1):

    text = template.format(
        INPUT=f"wr_trans{n-1}.trn",
        OUTPUT=f"wr_trans{n}.trn",
        CONT=f"last_cont_trans{n}.dat"
    )

    with open(f"model_{n}.in", "w") as f:
        f.write(text)
```

```python
cloudy = "/home/anwesha_barman/c25.00/source/cloudy.exe"

for n in range(1, N+1):

    print(f"\nRunning model_{n}.in")

    status = os.system(
        f"{cloudy} -r model_{n}"
    )

    if status != 0:
        print(f"Run {n} failed with status {status}")
        break

    print(f"Run {n} completed successfully")
```
