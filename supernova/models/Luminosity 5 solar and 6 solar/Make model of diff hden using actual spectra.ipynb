```python
with open("actual.in") as f:
    template = f.read()
hden = [8.8, 8.9,9.1, 9.5]
for h in hden:
    for L in range(6,7):

        text = template.replace("hden 8, power =-2",f"hden {h}, power =-2")

        text = text.replace("luminosity solar 5",f"luminosity solar {L}")

        text = text.replace('file="spectra_L6.dat"',f'file="spectra_h{h}_L{L}.dat"')

        with open(f"model_h{h}_L{L}.in", "w") as f:
            f.write(text)

print("All models generated.")
```
